```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=1.0)
```

```python
# Exemplo 1: Prompting tradicional, sem RAG

pergunta = "Qual é a política de home office da nossa empresa?"
prompt_tradicional = ChatPromptTemplate.from_template(
    f"Responda a seguinte a pergunta: {pergunta}"
)
```

```python
print(prompt_tradicional)
```

```python
chain_tradicional = prompt_tradicional | llm
resposta_tradicional = chain_tradicional.invoke({"pergunta": pergunta})
```

```python
print(resposta_tradicional.content)
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("../../data/politica_home_office.pdf")
documento = loader.load()
```

```python
documento
```

```python
contexto_empresa = documento[0].page_content
print(contexto_empresa)
```

```python
# Exemplo 2: Com RAG
prompt_rag = ChatPromptTemplate.from_template(f"""
    Use o contexto abaixo para responder a pergunta.
    Se não souber a resposta baseado no contexto, diga que não tem informação.
    Contexto: {contexto_empresa}
    Pergunta: {pergunta}
    Resposta:""")
```

```python
chain_rag = prompt_rag | llm
resposta_rag = chain_rag.invoke({
    "contexto": contexto_empresa,
    "pergunta": pergunta})
```

```python
print(resposta_rag.content)
```
