```python
import os
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get('ALURA_AULA')
```

```python
from google import genai

client = genai.Client()
```

```python
resposta = client.models.generate_content(model="gemini-2.5-flash", contents="O que é IA?")
```

```python
print(resposta.text)
```

```python
chat = client.chats.create(model="gemini-2.5-flash")
```

```python
chat.get_history()
```

```python
respostas = []
pergunta = input("Digite a sua pergunta pro Gemini: ")

while pergunta != "fim":
  if len(pergunta) > 0:
    resposta = chat.send_message(pergunta).text
    respostas.append(resposta)
    print(resposta)
  pergunta = input("\n\nDigite a sua pergunta pro Gemini: ")
```
