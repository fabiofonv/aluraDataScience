```python
from dotenv import load_dotenv
load_dotenv(verbose=True)

PERSIST_DIRECTORY="./chroma_rh"
EMBEDDING_MODEL="gemini-embedding-001"
LLM_MODEL="gemini-2.5-flash"
```

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader

pdfs = PyPDFDirectoryLoader(
    "../data",
    glob="*23.pdf",
).load()
```

```python
from langchain_text_splitters import CharacterTextSplitter

token_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    encoding_name="cl100k_base",
    chunk_size=1000,
    chunk_overlap=100
)

token_chunks = token_splitter.split_documents(pdfs)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(
    model=EMBEDDING_MODEL
)
```

```python
splitter = CharacterTextSplitter.from_tiktoken_encoder(
    model_name="gpt-4",
    chunk_size=1000,
    chunk_overlap=100
)

chunks = splitter.split_documents(pdfs)
```

```python
from langchain_experimental.text_splitter import SemanticChunker

# O semantic chunker vai dividir textos em chunks que tenham congruência, vetores próximos
semantic_splitter = SemanticChunker(embeddings=embeddings)
semantic_chunks = semantic_splitter.split_documents(pdfs)
```
