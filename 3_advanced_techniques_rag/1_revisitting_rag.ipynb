````python
from dotenv import load_dotenv
load_dotenv(verbose=True)

PERSIST_DIRECTORY="./chroma_rh"
EMBEDDING_MODEL="gemini-embedding-001"
LLM_MODEL="gemini-2.5-flash"
````

````python
from langchain_community.document_loaders import TextLoader

document = TextLoader("../data/GTB_gold_Nov23.txt", encoding="utf-8").load()
````

````python
document
````

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)

chunks = splitter.split_documents(document)
````

````python
len(chunks)
````

````python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(
    model=EMBEDDING_MODEL
)
````

````python
print(f"{chunks[0].page_content}\n\n")

embeddings.embed_query(chunks[0].page_content)
````

````python
from langchain_community.vectorstores import InMemoryVectorStore

vectorstore = InMemoryVectorStore.from_documents(
    documents=chunks,
    embedding=embeddings
)
````

````python
retriever = vectorstore.as_retriever(
    search_kwargs={"k": 2}
)
````

````python
retriever.invoke("Seguro viagem")
````

````python
query = "Como devo proceder caso tenha um item comprado roubado?"

query_embed = embeddings.embed_query(query)
````

````python
query_embed
````

````python
similar_chunks = retriever.invoke(query)
````

````python
similar_chunks
````

````python
similar_texts = [chunk.page_content for chunk in similar_chunks]
````

````python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [("system", "Responda usando exclusivamente o conteúdo fornecido.\n\nContexto:{context}"),
     ("human", "{query}")]
)
````

````python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser

llm = ChatGoogleGenerativeAI(
    model=LLM_MODEL,
    temperature=0.2
)
````

````python
llm.invoke(query)
````

````python
chain = prompt | llm | StrOutputParser()

pieces = retriever.invoke(query)

context = "\n\n".join(piece.page_content for piece in pieces)

chain.invoke({
    "query": query,
    "context": context
})
````

````python
from langchain_core.globals import  set_debug
set_debug(False)
chain.invoke({
    "query": query,
    "context": context
})
````

# 🐍 Python Virtual Environments: Activation and Shell Differences

When setting up a Python environment, activating the virtual environment is a critical step to ensure that packages are installed in isolation, preventing dependency conflicts between projects or with the global system.

---

## 🖥️ Differences Between Shells

While the goal is the same, activation methods vary significantly across operating systems and terminals:

* **Unix-based (Linux/macOS):** Terminals like **Bash** or **Zsh** use the `source` command to read the activation script and update environment variables.
* **Windows PowerShell:** Often enforces strict **Execution Policies**. You might encounter an "unauthorized" error because PowerShell restricts running scripts by default to protect the system.
* **Windows Command Prompt (CMD):** Generally has fewer restrictions than PowerShell. Developers often switch to CMD to bypass PowerShell permission issues using the `activate.bat` file.



---

## 🛠️ Activation Approaches by Shell

Use the following commands based on your specific environment (assuming your environment folder is named `env`):

### 🐧 Linux / macOS (Bash/Zsh)
```bash
source env/bin/activate

```

### 🪟 Windows Command Prompt (CMD)

```cmd
env\Scripts\activate.bat

```

### 🟦 Windows PowerShell

```powershell
.\env\Scripts\Activate.ps1

```

> **Note:** If you see a permission error, you may need to adjust your execution policy. Use `Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser` with caution.

---

## 🛡️ Security Considerations

Execution policies exist to prevent the inadvertent running of malicious scripts. When adjusting these settings (especially `Set-ExecutionPolicy`), it is vital to:

1. **Understand the Scope:** Only change policies for the `CurrentUser` or the specific process whenever possible.
2. **Verify Sources:** Only run activation scripts generated by trusted tools like `venv`, `virtualenv`, or `conda`.
3. **Maintain Best Practices:** Keeping a secure development environment protects both your machine and your ongoing projects from external vulnerabilities.

---

> **Key Takeaway:** Modularizing your dependencies through virtual environments is the first step toward a professional and reproducible Data Science or Machine Learning workflow.
